```scala
val haiRaw = spark.read.option("header", "true").option("inferSchema", "true").csv("/WideWorldImporters/transactionDetail.txt")
```

```scala
haiRaw.printSchema()
```

```scala
haiRaw.cache()
```

```scala
haiRaw.show(1)
```

```scala
haiRaw.registerTempTable("transactiondetail")
```

```scala
%%sql
SELECT Description,
    Quantity,
    UnitPrice,
    Profit
FROM transactiondetail
```

```scala
%%sql
SELECT Customer,
    BuyingGroup,
    SUM(Quantity) AS Quantity,
    SUM(Profit) AS Profit,
    SUM(TotalIncludingTax) AS TotalIncludingTax,
    SUM(TotalDryItems) AS TotalDryItems,
    SUM(TotalChillerItems) AS TotalChillerItems
FROM transactiondetail
GROUP BY Customer, BuyingGroup
```

```scala
val cust = spark.sql("""SELECT Customer,
                         BuyingGroup,
                         SUM(Quantity) AS Quantity,
                         SUM(Profit) AS Profit,
                         SUM(TotalIncludingTax) AS TotalIncludingTax,
                         SUM(TotalDryItems) AS TotalDryItems,
                         SUM(TotalChillerItems) AS TotalChillerItems
                        FROM transactiondetail
                        GROUP BY Customer, BuyingGroup""")
cust.write.mode("overwrite").format("parquet").saveAsTable("customersummary")
```
